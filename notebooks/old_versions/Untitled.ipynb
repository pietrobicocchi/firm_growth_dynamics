```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


# Define the ticker symbols and timeframe
tickers = ["AAPL", "GOOG", "AMZN", "MSFT", "JPM", "WMT", "JNJ", "PG", "XOM", "TSLA"]
start_date = "2010-01-01"
end_date = "2022-05-10"

# Create an empty DataFrame to store the data
data = pd.DataFrame()

# Loop through each ticker symbol and download the data
for ticker in tickers:
    ticker_data = yf.download(ticker, start=start_date, end=end_date)
    ticker_data['ticker'] = ticker
    data = pd.concat([data, ticker_data], axis=0)

# Reset the index and display the data
data.reset_index(inplace=True)
print(data.head())
```

```python
from matplotlib import style
style.use('dark_background')

# Define the ticker symbol and timeframe
ticker = "AAPL"
start_date = "2010-01-01"
end_date = "2022-05-10"

# Import the data and calculate returns
data = yf.download(ticker, start=start_date, end=end_date)
data['returns'] = data['Adj Close'].pct_change()

# Calculate the 50-day and 200-day moving averages
data['MA50'] = data['Adj Close'].rolling(window=50).mean()
data['MA200'] = data['Adj Close'].rolling(window=200).mean()

# Plot the data and moving averages
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Adj Close'], label='Price')
plt.plot(data.index, data['MA50'], label='50-day MA')
plt.plot(data.index, data['MA200'], label='200-day MA')
plt.legend()
plt.title(ticker)
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
